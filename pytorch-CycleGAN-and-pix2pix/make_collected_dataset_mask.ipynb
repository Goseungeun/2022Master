```python
from PIL import Image
import os
import numpy as np
import matplotlib.pylab as plt
```

```python
mask_image_path = 'D:/Dataset/20230821_soiling/SegmentationClass'
save_path = 'D:/Dataset/20230821_soiling/Mask'
mask_images =os.listdir(mask_image_path)

for mask in mask_images:
    _img = Image.open(os.path.join(mask_image_path,mask)).convert('L')
    mask,ext = os.path.splitext(mask)
    m = np.array(_img)
    for i in m:
        for j in range(len(i)):
            if i[j] == 13:
                i[j] = 0
            else:
                i[j] = 255
    mask_img = Image.fromarray(m)
    mask_img.save(os.path.join(save_path,mask+'.jpg'))
```
